```python
# Import essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Load dataset
df = pd.read_csv('../data/EV_Battery_Performance.csv')
```

```python
# Display first few rows
df.index = range(1, len(df) + 1)
df.head()
```

```python
# Get basic info about the dataset
df.info()
```

```python
# Shape of the dataset
print("Shape of dataset:", df.shape)
```

```python
# Check column names
print("\nColumns in dataset:\n", df.columns.tolist())
```

```python
# Descriptive statistics
df.describe().round(2)
```

```python
# Missing values
print("\nMissing values in each column:\n", df.isnull().sum())
```

```python
# Check for duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# Display unique counts for categorical features if any
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"{col}: {df[col].nunique()} unique values")
```

```python
# Range Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['range_km'], kde=True, color='green')
plt.title("Distribution of EV Range (km)")
plt.xlabel("Range (km)")
plt.show()
```

```python
# Battery Capacity vs Range
plt.figure(figsize=(8,5))
sns.scatterplot(x='battery_capacity_kWh', y='range_km', data=df, hue='efficiency_wh_per_km', palette='viridis')
plt.title("Battery Capacity vs Range (Colored by Efficiency)")
plt.show()
```

```python
# Efficiency Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['efficiency_wh_per_km'], kde=True, color='orange')
plt.title("Efficiency Distribution (Wh/km)")
plt.show()
```

```python
# Compute correlation matrix
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of EV Features")
plt.show()
```

### Observations:
- The dataset contains around *N* records and *M* features.  
- **Range** shows strong correlation with **Battery Capacity (kWh)**.  
- Efficiency inversely affects range — higher Wh/km → lower range.  
- Data appears mostly clean with minimal missing values.  
- These insights will guide model selection in Week 2 for range/battery prediction.
